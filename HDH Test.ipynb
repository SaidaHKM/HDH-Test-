### Table Car Site  (SQLITE) au format OMOP-CDM

```python
#Connexion a une base de données sqlite 
connection_obj = sqlite3.connect('mydb.db')
#Creation d'une connection : cursor object
cursor_obj = connection_obj.cursor()
```

```python
#Création de la table Care_Site et etablissement_sante 

Table1="""CREATE TABLE IF NOT EXISTS Care_Site (
                care_site_id INTEGER PRIMARY KEY, 
                care_site_name TEXT, 
                place_of_service_concept_id INTEGER, 
                location_id INTEGER,
                care_site_source_value TEXT,
                place_of_service_source_value TEXT
        );"""
    

Table2="""CREATE TABLE IF NOT EXISTS etablissement_sante (
                eta_num INTEGER PRIMARY KEY, 
                soc_rai TEXT
        );"""
     
cursor_obj.execute(Table1)                  
cursor_obj.execute(Table2)                

#Insertion des données sur les établissements dans la table etablissement_sante
cursor_obj.execute('''INSERT INTO etablissement_sante (eta_num, soc_rai)
VALUES ('750300360', 'lHôpital Privé des Peupliers'),('750023772', 'Pharmacie Plaisance');''')
```

### Table Car Site au format OMOP-CDM

```python
##Creation de la table provider 

Car_site = pd.DataFrame({
    "care_site_id" : [],
    "care_site_name"   :  [],
    "place_of_service_concept_id"   :  [],
    "location_id"   :  [],
    "care_site_source_value"   :  [],
    "place_of_service_source_value"   :  []
       
    
})
Car_site
```

```python
Car_site["care_site_id"]=etablissement_sante["eta_num"]
Car_site["care_site_name"]=etablissement_sante["soc_rai"]

Car_site
```

### Table Provider au format OMOP-CDM

```python
#Creation de la table provider 

Provider = pd.DataFrame({
    "provider_id" : [],
    "provider_name"   :  [],
     "npi"   :  [],
     "dea"   :  [],
     "specialty_concept_id"   :  [],
     "care_site_id"   :  [],
     "year_of_birth"   :  [],
     "gender_concept_id"   :  [],
     "provider_source_value"   :  [],
     "specialty_source_value"   :  [],
     "specialty_source_concept_id"   :  [],
     "gender_source_value"   :  [],
     "gender_source_concept_id"   :  []  
    
})
Provider
```

```python
#Insertion des données via les tables professionnels_act et professionnels_spe

Provider["provider_id"]=pd.concat([professionnels_act["pfs_act_nat"], professionnels_spe["pfs_spe_cod"]],ignore_index=True)
Provider["specialty_source_value"]=pd.concat([professionnels_act["label"], professionnels_spe["label"]],ignore_index=True)
```

### Table sur les prestations au format OMOP-CDM

```python
infos_pres_remb
```

```python
# Insertion du nom de l'établissement où le soin a été réalisé

infos_pres_remb_CDM=pd.merge(infos_pres_remb,Car_site[['care_site_id','care_site_name']],left_on="etb_pre_fin", right_on="care_site_id",how="outer")
infos_pres_remb_CDM=infos_pres_remb_CDM.drop('care_site_id',axis=1)
```

```python
# Insertion de la nature du soin réalisé

infos_pres_remb_CDM=pd.merge(infos_pres_remb_CDM,Provider[['provider_id','specialty_source_value']],left_on="pse_act_nat", right_on="provider_id",how="left")
infos_pres_remb_CDM=infos_pres_remb_CDM.drop('provider_id',axis=1)
infos_pres_remb_CDM.rename(columns={'specialty_source_value_x': 'specialty_source_value_act'}, inplace=True)
```

```python
# Insertion de la spécialité du soin réalisé

infos_pres_remb_CDM=pd.merge(infos_pres_remb_CDM,Provider[['provider_id','specialty_source_value']],left_on="pse_spe_cod", right_on="provider_id",how="outer")
infos_pres_remb_CDM=infos_pres_remb_CDM.drop('provider_id',axis=1)
infos_pres_remb_CDM.rename(columns={'specialty_source_value_y': 'specialty_source_value_spe'}, inplace=True)
```
